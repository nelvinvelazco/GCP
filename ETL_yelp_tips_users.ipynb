#### User Yelp

```python
df_users_yelp = pd.read_parquet('C:/Data/Yelp/user/user_part_10.parquet')
df_users_yelp.head(3)
```

##### Dividir el Archvivo "user.parquet" en 10 partes

```python
num_filas = len(df_users_yelp)
filas_X_parte = num_filas // 10

# Dividir y guardar cada parte
for i in range(10):
    indice_inicio = i * filas_X_parte
    if i == 9:
        indice_final = num_filas    # Para que la ultima parte incluya los registros restantes
    else:
        indice_final = (i + 1) * filas_X_parte
    
    # Seleccionar la parte del DataFrame
    df_user_part = df_users_yelp.iloc[indice_inicio:indice_final]
    
    # Guardar la parte en un nuevo archivo Parquet
    df_user_part.to_parquet(f'user_part_{i+1}.parquet')
```

```python
df_users_yelp= df_users_yelp.drop(['elite', 'friends', 'compliment_hot', 'compliment_more', 'compliment_profile', 'compliment_cute', 'compliment_list',
                                   'compliment_note', 'compliment_plain', 'compliment_cool', 'compliment_funny', 'compliment_writer','compliment_photos'], axis=1)       #Elimina la columnas
```
